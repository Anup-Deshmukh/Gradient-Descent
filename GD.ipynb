## Load modules

```python
# !pip install numpy, scipy, scikit-image, skimage, matplotlib

import matplotlib.pyplot as plt

from skimage.color import rgb2gray
from skimage import data
from skimage.transform import resize

# Numpy is useful for handling arrays and matrices.
import numpy as np
```

## Load image

```python
img = data.astronaut()
img = rgb2gray(img)*255 # convert to gray and change scale from (0,1) to (0,255).
n = 512
plt.figure(1, figsize=(10, 10))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## Computing the differences operators here.

```python
# You will need these three methods to construct sparse differences operators.
# If you do not use sparse operators you might have scalability problems.
from scipy import sparse
from scipy.sparse import diags
from scipy.sparse import kron
from scipy.sparse import identity


# Start coding here.
J = sparse.csr_matrix(-identity(n) + diags([1] * (n-1), 1))

I = sparse.csr_matrix(identity(n))

D_h = sparse.csr_matrix(kron(J, I))

D_v = sparse.csr_matrix(kron(I, J))

D = sparse.csr_matrix(D_h + D_v*1j)
D_star = sparse.csr_matrix(np.conj(D).transpose())

print(D_h.shape)
print(D_v.shape)
```

## Add noise to the image

```python
mean_ = 0
standard_deviation = 30
dimensions = (n,n)

noise = np.random.normal(mean_,standard_deviation,dimensions)

noisy_image = img + noise

plt.figure(1, figsize=(10, 10))
plt.imshow(noisy_image, cmap='gray', vmin=0, vmax=255)
plt.show()
```

## Here I implement gradient descent using the Lipschitz constant as the step-size for the denoising problem. I use eigsh method from scipy.sparse.linalg to compute the Lipschitz constant.


```python
from scipy.sparse.linalg import eigsh
from numpy import linalg as LA
import time

def gradient_descent(x0, epsilon, lambda_, max_iterations):
    
# x0: is the initial guess for the x variables
# epsilon: is the termination tolerance parameter
# lambda_: is the regularization parameter of the denoising problem.
# max_iterations: is the maximum number of iterations that you allow the algorithm to run.


    l_s_time = time.time()
    
    D_h_tran = sparse.csr_matrix(D_h).transpose()
    D_v_tran = sparse.csr_matrix(D_v).transpose()

    term1 = np.dot(D_h_tran, D_h)
    term2 = np.dot(D_v_tran, D_v)
    Dconj_D = sparse.csr_matrix(term1 + term2)

    identity_ls = identity(n*n)
   
    mat = sparse.csr_matrix(lambda_*(Dconj_D) + identity_ls).tocsr()
    matconj = sparse.csr_matrix(np.conj(mat).transpose())
    matconj_mat = matconj.dot(mat)
  
    eigenvalues = eigsh(matconj_mat, return_eigenvectors = False) #calculating Lipschitz constant
    print(eigenvalues)
    max_ev = np.amax(np.array(eigenvalues)) #taking the max eigenvalue
    
    lip_constant = np.sqrt(max_ev) #Lipschitz constant
    
    lip_constant = 32.999
    l_e_time = time.time()
    print(lip_constant)
    
    l_time_taken = l_e_time - l_s_time
    print("time taken to compute Lipschitz constant is ", l_time_taken, "seconds") 
    
    ### gradient descent with constant step sizes ### 
    grad_s_time = time.time()
    norm_grad = float("inf")
    num_iter = 0
    z_n = x0    
    x = x0
    grad_x = lambda_*(Dconj_D.dot(x)) + x - z_n
    f_x = np.zeros(max_iterations)    
    D = sparse.csr_matrix(D_h + D_v*1j)
    f_x[0] = lambda_/2*LA.norm(D.dot(x))**2 + 1/2*LA.norm(x-z_n)**2
    norm_grad = LA.norm(grad_x)    
    i=1
    while norm_grad > epsilon and num_iter < max_iterations:            
        h = -grad_x/lip_constant
        x = x + h # updating x 
        grad_x = lambda_*(Dconj_D.dot(x)) + x - z_n # calculating the gradient of x
        norm_grad = LA.norm(grad_x)
        f_x[i] = lambda_/2*LA.norm(D.dot(x))**2 + 1/2*LA.norm(x-z_n)**2 # the objective function at x
        num_iter = num_iter + 1
        i = i+1
        
    print("number of iterations required to converge = ", num_iter) 
    grad_e_time = time.time()
    grad_time_taken = grad_e_time - grad_s_time
    print("time taken gradient descent is ", grad_time_taken, "seconds") 
    
    return f_x #returing objective function of denoising problem for every x_k 
    
```

## Call Gradient Descent

```python
# Initialize parameters of gradient descent.
lambda_ = 4
epsilon = 1.0e-2
max_iterations = 2000

# Set x0 equal to the vectorized noisy image.
vec_noisy_image = noisy_image.flatten('F')
f_x = gradient_descent(vec_noisy_image, epsilon, lambda_, max_iterations)
```

## Plot of  $$(f(x_k) - f(x^*)) / (f(x_0) - f(x^*))$$ vs the iteration counter k, where $$x^*$$ is the minimizer of the denoising problem, which I compute by using spsolve.

```python
# Plot the rellative objective function vs number of iterations. 
# Write your code here.
# Here is an example code 


### clculating minimizer of the denoising problem using spsolve ###
D = sparse.csr_matrix(D_h + D_v*1j)
D_h_tran = D_h.transpose()
D_v_tran = D_v.transpose()
identity_ls = identity(n*n)

term1 = np.dot(D_h_tran, D_h)
term2 = np.dot(D_v_tran, D_v)
Dconj_D = lambda_*(term1 + term2)
A_ls = Dconj_D + identity_ls

vec_noisy_image = noisy_image.flatten('F')
B_ls = sparse.csr_matrix(vec_noisy_image).transpose()

denoised_image = sparse.linalg.spsolve(A_ls, B_ls) #spsolve
denoised_image = sparse.csr_matrix(denoised_image).transpose()

x_star = sparse.csr_matrix(denoised_image).tocsr() #denoised image
x_star = x_star.toarray().flatten('F')
f_x_star = lambda_/2*LA.norm(D.dot(x_star))**2 + 1/2*LA.norm(x_star-vec_noisy_image)**2
### calculating minimizer of the denoising problem using spsolve ###

print(f_x.shape[0])
### Q2: Plotting the theoretical convergence rate ###
k = np.arange(1,f_x.shape[0]+1)
theory_rate = 1-1/32.999   # 1-mu/L where mu = 1 as proven in Q7 and L = 32.999 which is calculated in Q1
f_theory = np.zeros(max_iterations)
iter = 0    
for i in k:
    f_theory[iter] = theory_rate**i
    iter = iter + 1   
### Q2: Plotting the theoretical convergence rate ###

f_theory = np.trim_zeros(f_theory)    
f_x = np.trim_zeros(f_x)
y_axis_term = (f_x - f_x_star)/(f_x[0] - f_x_star)

fig = plt.figure(figsize=(8, 6))
plt.plot(k, y_axis_term, label=("Gradient descent + Lipschitz"), linewidth=5.0, color ="black")
plt.plot(k, f_theory, label=("Theoretical convergence rate"), linewidth=5.0, color ="red")

plt.legend(prop={'size': 20},loc="upper right")
plt.xlabel("iteration $k$", fontsize=25)
plt.ylabel("Relative distance to opt.", fontsize=25)
plt.grid(linestyle='dashed')
plt.show()
```
